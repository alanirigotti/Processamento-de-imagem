# Exemplo 03 – Abrir uma imagem colorida em RGB, visualizar e salvar cada um dos canais separadamente. Obs: Busquem compreender o que significa cada um dos canais.

Em cada pixel de uma imagem é composto por 3 componentes de 8 bits cada. Sendo assim, cada canal é uma imagem em escala Red, Blue e Green que representa a intensidade de uma das cores no espaço RGB. Basicamente, cada pixel (ponto da imagem) é uma mistura dessas cores.

O output é uma imagem para cada canal, onde cada imagem destaca uma das cores primárias Assim, os canais possuem valores de 0 a 255, dependendo a intensidade da cor do canal na imagem.

```python
# Abrir a imagem em RGB
imagem = cv2.imread('LenaC.jpg')  # lê a imagem no formato BGR
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)  # converter para RGB

 # separa os canais R, G e B com split
r, g, b = cv2.split(imagem_rgb)

# visualizar os canais separadamente
plt.figure(figsize=(10, 10))

# Canal Vermelho
plt.subplot(1, 3, 1)
plt.imshow(r, cmap='Reds')
plt.title('Canal Vermelho')
plt.axis('off')

# Canal Verde
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
plt.title('Canal Verde')
plt.axis('off')

# Canal Azul
plt.subplot(1, 3, 3)
plt.imshow(b, cmap='Blues')
plt.title('Canal Azul')
plt.axis('off')

plt.show()

# salvar as imagens dos canais separadamente no collab
cv2.imwrite('canal_vermelho.jpg', r)
cv2.imwrite('canal_verde.jpg', g)
cv2.imwrite('canal_azul.jpg', b)

# salvar a imagem RGB original
cv2.imwrite('imagem_rgb.jpg', cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2BGR))
```

Apliquem os filtros passa baixa mediana (median) e media (blur), visualizem os resultados e salvem. Obs: busquem compreender os resultados de cada filtro.

```python
# filtro passa-baixa mediana - substitui cada pixel pelo valor mediano dos vizinhos, preservando melhor as bordasda imagem e removendo ruídos especificos
imagem_mediana = cv2.medianBlur(imagem_gray, 5)

# Mostra a imagem filtrada
cv2_imshow(imagem_mediana)

# Salva a imagem filtrada
cv2.imwrite("saida_mediana.jpg", imagem_mediana)
```

```python
# aplica o filtro média - blur blur - suaviza a imagem calculando a média dos pixels vizinhos, causa um efeito de borrado na imagem
imagem_media = cv2.blur(imagem_gray, (5, 5))

# mostra a imagem filtrada
cv2_imshow(imagem_media)

 # Salva a imagem filtrada
cv2.imwrite("saida_media.jpg", imagem_media)
```

# Exercício 5

Abrir uma imagem colorida, transformar em tom de cinza, visualizar imagem de entrada. Apliquem uma limiarização (thresholding), visualizem os resultados e salvem. Obs: busquem compreender os resultados da técnica.

```python
# Abrir a imagem colorida
imagem_colorida = cv2.imread('LenaC.jpg')

# Transformar a imagem em tons de cinza (escala de cinza)
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY) # poderia usar a imagem no tom cinza feita anteriormente

# Exibir a imagem original e a imagem em tons de cinza
plt.figure(figsize=(12, 6))

# Exibir a imagem colorida
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')
plt.axis('off')

# Exibir a imagem em tons de cinza
plt.subplot(1, 2, 2)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')

plt.show()

# Aplicar limiarização (thresholding) usando o valor 127 (limiar)- o qual o valor é onde os pixels abaixo vão se tornar 0 (preto) e os acima se tornam 255 (branco)
_, imagem_limiarizada = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Exibir a imagem limiarizada
plt.figure(figsize=(6, 6))
plt.imshow(imagem_limiarizada, cmap='gray')
plt.title('Imagem Limiarizada (Thresholding)')
plt.axis('off')
plt.show()
```

```python
# Salvar a imagem limiarizada
cv2.imwrite('imagem_limiarizada.jpg', imagem_limiarizada)
```

# Exercício 6

Abrir uma imagem colorida, transformar para tom de cinza e aplicar uma Equalização de histograma utilizando a OpenCv, visualizando a imagem de entrada e seu respectivo histograma inicialmente, e, em seguida, o resultado da equalização e seu histograma. Esta técnica aumenta o contraste da imagem

```python
# 1. Abrir a imagem colorida
imagem_colorida = cv2.imread('LenaC.jpg')
#imagem_colorida = cv2.imread('Bola.jpg')
#imagem_colorida = cv2.imread('LenaPB.jpg')
```

```python
# transforma a imagem colorida para tons de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)
```

```python
# exibe a imagem original em tons de cinza
plt.figure(figsize=(12, 6))

# mostra a imagem original em tons de cinza
plt.subplot(2, 2, 1)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original em Tons de Cinza')
plt.axis('off')
```

```python
# exibe o histograma da imagem original
plt.subplot(2, 2, 2)
plt.hist(imagem_cinza.ravel(), bins=256, range=[0, 256], color='black')
plt.title('Histograma da Imagem Original')
```

```python
# aplica a equalização de histograma para distribuir a luminosidade da imagem
imagem_equalizada = cv2.equalizeHist(imagem_cinza)
```

```python
# exibe a imagem equalizada
plt.subplot(2, 2, 3)
plt.imshow(imagem_equalizada, cmap='gray')
plt.title('Imagem Equalizada')
plt.axis('off')
```

```python
# exibe o histograma da imagem equalizada
plt.subplot(2, 2, 4)
plt.hist(imagem_equalizada.ravel(), bins=256, range=[0, 256], color='black')
plt.title('Histograma da Imagem Equalizada')
```

```python
# exibe todas as imagens e gráficos
plt.tight_layout()
plt.show()
```

```python
# salvar a imagem equalizada no collab
cv2.imwrite('imagem_equalizada.jpg', imagem_equalizada)
```
